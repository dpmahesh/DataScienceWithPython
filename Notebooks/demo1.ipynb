```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =sns.load_dataset('tips')
df
```

```python
df.hist(bins=10,figsize=(10,5))
plt.show()
```

```python
sns.displot(data=df,x='day',y='tip',hue='sex')
plt.show()
```

```python
sns.boxenplot(data=df,x='day',y='tip',hue='sex',dodge=False)
plt.show()
```

```python
sns.boxplot(data=df,x='day',y='tip',hue='sex',dodge=False)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('planets')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
sns.lineplot(data=df,x='year',y='mass')
plt.show()
```

```python
plt.figure(figsize=(15,6))
ax = sns.countplot(data=df,x='method',order=df['method'].value_counts().index)
ax.bar_label(ax.containers[0])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure().set_figwidth(15)
ax=sns.countplot(data=df,x='year')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
sns.scatterplot(data=df,x='mass',y='distance')
plt.show()
```

```python
sns.scatterplot(data=df,x='orbital_period',y='distance',hue='mass')
plt.show()
```
